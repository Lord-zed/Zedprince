### Measure of centre

picking some selected variable to carry out our analysis

```python
# picking our selected Numerica variables
selected_variables = ['hdlngth', 'totlngth', 'taill', 'footlgth']
```

```python
# Calculate measures of center for key variables
measures_center = pd.DataFrame()
for var in selected_variables:
    measures_center[var] = {
        'Mean': df[var].mean(),
        'Median': df[var].median(),
        'Mode': df[var].mode().iloc[0] if not df[var].mode().empty else np.nan
    }
    print(measures_center.round(2))
```

| totlngth| 87.09 | 88.00  | 4.31  | 89.00  | The mean is slightly lower than the median. This indicates a very slight left-skew (a few smaller possums are pulling the mean down), but the typical total length is around 88 mm.            |
| taill   |  37.01| 37.00  | 1.96  | 38.00  | The mean is slightly higher than the median. This indicates a slight right-skew (a few possums with longer tails are pulling the mean up). The typical tail length is about 36.5 cm.            |
|footlngth| 67.84 | 67.95  | 4.40  | 73.20  | The mean and median are almost identical. This points to a very symmetrical distribution of foot lengths among the possums. The typical foot length is about 68.5 cm.            |

### Measure of Spread

```python
# calculating measures of spread for our key variables
measures_spread = pd.DataFrame()

for var in selected_variables:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    
    measures_spread[var] = {
        'Range': df[var].max() - df[var].min(),
        'Variance': df[var].var(),
        'Standard Deviation': df[var].std(),
        'IQR': IQR,
        'Q1 (25th percentile)': Q1,
        'Q3 (75th percentile)': Q3,
        'Coefficient of Variation': (df[var].std() / df[var].mean()) * 100,
        'skewness': df[var].skew(),
        'kurtosis' :df[var].kurtosis()
    }
print(measures_spread.round(2))
```

### Working on outlier

```python
# working out the outlier
outliers = {}

for var in selected_variables:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_boundary = Q1 - 1.5 * IQR
    upper_boundary = Q3 + 1.5 * IQR
    
    # Find outliers
    lower_outliers = df[df[var] < lower_boundary]
    upper_outliers = df[df[var] > upper_boundary]
    
    outliers[var] = {
        'Lower boundary': lower_boundary,
        'Upper boundary': upper_boundary,
        'Number of Lower Outliers': len(lower_outliers),
        'Number of Upper Outliers': len(upper_outliers),
        'Total Outliers': len(lower_outliers) + len(upper_outliers),
        'Lower Outlier Cases': list(lower_outliers['case'].values) if len(lower_outliers) > 0 else [],
        'Upper Outlier Cases': list(upper_outliers['case'].values) if len(upper_outliers) > 0 else []
    }
    
    print(f"\n--- {var} ---")
    print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"Lower boundary: {lower_boundary:.2f}, Upper boundary: {upper_boundary:.2f}")
    print(f"Lower Outliers: {len(lower_outliers)} cases")
    print(f"Upper Outliers: {len(upper_outliers)} cases")
    
    if len(lower_outliers) > 0:
        print(f"Lower outlier cases: {list(lower_outliers['case'].values)}")
        print("Details of lower outliers:")
        for col in ['case', 'Pop', 'sex', 'age', var]:
            if col in lower_outliers.columns:
                print(f"  {col}: {list(lower_outliers[col].values)}")
    
    if len(upper_outliers) > 0:
        print(f"Upper outlier cases: {list(upper_outliers['case'].values)}")
        print("Details of upper outliers:")
        for col in ['case', 'Pop', 'sex', 'age', var]:
            if col in upper_outliers.columns:
                print(f"  {col}: {list(upper_outliers[col].values)}")
```
